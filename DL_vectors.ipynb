```
plt.quiver(0,0,6,7, scale_units='xy', angles= 'xy', scale =1, color= 'r')
plt.quiver(0,0,-6,7, scale_units='xy', angles= 'xy', scale =1, color= 'g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
def plot_vect(x1,y1,x2,y2,c='b'):
  plt.quiver(x1,y1,x2,y2, scale_units='xy', angles= 'xy', scale =1, color= c)
  if abs(x1) > abs(x2):
    xlim = abs(x1) + 3
  else:
    xlim = abs(x2) + 3
  if abs(y1) > abs(y2):
    ylim = abs(y1) + 3
  else:
    ylim = abs(y2) + 3
  plt.xlim(-1*xlim,xlim)
  plt.ylim(-1*ylim,ylim)
  plt.show()
```

```
plot_vect(0,0,3,4, 'r')
```

```
plot_vect(0,0,5,10)
```

```
def plot_vect_1(vecs):
  colors = ['r','y','b','g']
  i=0
  for vec in vecs:
    plt.quiver(vec[0], vec[1], vec[2], vec[3], scale_units='xy', angles= 'xy', scale =1, color= colors[i%len(colors)])
    i+=1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
plot_vect_1([(0,0,5,7), (0,0,-4,9), (0,0,-2,-9), (0,0,9,-5)])
```

#Dot product

```
vecs =[np.asarray([0,0,3,4]), np.asarray([0,0,-3,4]), np.asarray([0,0,-3,-2]), np.asarray([0,0,4,-1])]
```

```
plot_vect_1(vecs)
```

```
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

$\vec{a}\cdot\vec{b} = |\vec{a}||\vec{b}|\cos\theta = a_xb_x + a_yb_y$

```
a_dot_b = np.dot(a,b)
```

```
print(a_dot_b)
```

$a_b = |\vec{a}|cos\theta = |\vec{a}|\frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}$

```
a_b = np.dot(a,b)/np.linalg.norm(b)
```

```
print(a_b)
```

$\vec{a_b} = a_b\hat{b} = a_b\frac{\vec{b}}{|\vec{b}|}$

```
vec_a_b = (a_b/np.linalg.norm(b))*b
```

```
print(vec_a_b)
```

```
type(vec_a_b)
```
